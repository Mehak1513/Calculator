```python
                                                                #Calculator
print('\t\t\t\t\t\t\t\t\tWELCOME\n\t\t\t\t\t\t\t\t    MEHAK CALCULATOR')
num1=int(input('Enter number'))
num2=int(input('Enter number'))
print('1) Addition\n2) Subtraction\n3) Multiply\n4) Divide')
selection= int(input('select the option'))
if selection==1:
    print('sum=',num1+num2)
elif selection==2:
    print('sub=',num1-num2)
elif selection==3:
    print('Multiply=',num1*num2)
elif selection==4:
    print('Division',num1/num2)

else:
    print('invalid selection')
```
